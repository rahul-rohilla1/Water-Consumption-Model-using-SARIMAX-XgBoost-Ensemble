```python
#convert rowkey to datetime
df['RowKey'] = pd.to_datetime(df['RowKey'])
```

```python
df["Sector_Neta"].unique()
```

```python
unique_sector_neta = [22120000, 22121000, 22130000, 22110000, 22131000, 22132100,22122000, 22122100]
```

```python
df['Canonical'].unique()
```

```python
unique_canonical = ['TOTALIZADOR', 'PRESION_ENTRADA_SECTOR', 'VOLUMEN_DIARIO', 'CAUDAL', 'PRESION_SALIDA_SECTOR', 'VOLUMEN']
```

```python
# Dictionary to store the filtered DataFrames
filtered_datasets = {}

# Loop through unique values of 'Sector_Neta' and 'Canonical' to filter the dataset
for sector in unique_sector_neta:
    for canonical in unique_canonical:
        key = f"sector_{sector}_canonical_{canonical}"
        filtered_datasets[key] = df[(df['Sector_Neta'] == sector) & (df['Canonical'] == canonical)]

# Display the keys for the filtered datasets to provide an overview of the splits
list(filtered_datasets.keys())
```

```python
zero_row_dict = {}
non_zero_row_dict = {}
for key in filtered_datasets:
    if filtered_datasets[key].shape[0] ==0:
        zero_row_dict[key] = filtered_datasets[key]
    else:
        non_zero_row_dict[key] = filtered_datasets[key]
    
        
```

```python
# Directory to save the CSV files
output_directory_series = "data_series"

# Create the directory if it doesn't exist
if not os.path.exists(output_directory_series):
    os.makedirs(output_directory_series)

# Save the filtered datasets to CSV again
for key, data in filtered_datasets.items():
    # Retain only the specified columns
    data = data[['RowKey', 'INF_Value']]
    
    # Save to CSV
    filename = f"{key}.csv"
    filepath = output_directory_series + filename
    data.to_csv(filepath, index=False)

output_directory_series
```
